```python
# necessary libraries
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
import math
import time
from datetime import datetime
```

```python
# sending a request to the website
base_url = "https://www.screener.in/screens/2500816/all-companies/"
response = requests.get(base_url)
# initializing a soup object
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
# getting the number of pages 
n_pages = math.ceil(int(soup.find_all('div', class_='options')[0]
              .find_all('a', class_='ink-900')[-1].text))
```
