```python
# necessary libraries
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
import math
```

```python
# sending a request to the website
base_url = "https://www.screener.in/screens/357649/all-listed-companies/?limit=25"
response = requests.get(base_url)
# initializing a soup object
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
# getting the number of pages 
n_pages = math.ceil(int(soup.find_all('div', class_='options')[0]
              .find_all('a', class_='ink-900')[-1].text))
```

```python
# dictionary to store the stock names tickers and links
stock_dict = {'stock_name': [], 'stock_ticker': [], 'stock_link': []}
for i in n_pages:
    url = base_url + '&page=' + str(i+1)'
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
```
